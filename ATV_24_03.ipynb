```
#Conversão de Horas=
#Escreva um programa que: Crie uma variável para armazenar uma hora inteira. Converta essa hora para minutos e segundos.

hora= int (input( "Informe uma hora inteira:"))
min= hora*60
seg= hora*3600
print (f"(hora) hora(s) equivale a (min) minutos e (seg) segundo")
```

```
#Cálculo de Troco=
# Escreva um programa que: Leia o valor pago por um produto, leia o preço do produto e que calcule e mostre o troco a ser devolvido.

valor_pago= float(input("Informe o valor pago:"))
preço_produto= float(input("Informe o preço do produto:"))
troco= valor_pago - preço_produto

print(f"O troco a ser devolvido é {troco}")
```

```
#Quociente e Resto=
# Escreva um programa que: Leia dois numeros inteiros e mostre o quociente e o resto da divisão intreira entre eles.

num1= int(input("Informe o primeiro numero:"))
num2= int(input("Informe o segundo numero:"))

quociente= num1//num2
resto= num1 % num2
print(f"O quociente é {quociente}")
print(f"O resto é {resto}")
```

```
#Salário com Comissão=
# Escreva um programa que: Leia o salário fixo de um vendedor, leia o total de vendas feitas por ele, leia a porcentagem de comissão e que calcule e mostre a comissão e salário final.

salario_fixo= float(input("Informe o salário fixo:"))
total_vendas= float(input("Informe o total de vendas:"))
porcentagem= float (input("Informe a porcentagem de comissão"))

comissão= total_vendas*(porcentagem/100)
salario_final= salario_fixo + comissão
print(f"A comissão é {comissão}")
print(f"O salário final é {salario_final}")
```

```
#Média Artimética=
# Escreva um programa que: Leia tres notas de um aluno e que calcule e mostre a média aritmética.

nota1 =float(input("Informe a primeira nota:"))
nota2 =float(input("Informe a segunda nota:"))
nota3 =float(input("Informe a terceira nota:"))

media= (nota1 + nota2 + nota3) /3
print(f"A media é {media}")

```

```
#Gasolina no tanque=
# Escreva um programa que: Leia o preço do litro da gasolina, leia o quanto o motorista quer pagar e que calcule e mostre quantos litros ele consegue colocar.

preço_litro= float(input("Informe o preço da gasolina:"))
valor_gastar= float(input("Informe o valor a ser gasto:"))

litros= valor_gastar/preço_litro
print(f"Voce pode colocar {litros}litros")

```

```
#Restaurante a Quilo=
# Escreva um programa que: Leia o peso do prato(em gramas) e calcule o valor a pagar, sabendo que cada quilo custa R$12,00.

peso_gramas= float(input("Informe o peso do prato em gramas:"))
peso_quilo= peso_gramas/1000
valor= peso_quilo*12

print(f"O valor a pagar é {valor}")
```

```
#Percentual de Votos=
# Escreva um programa que: Leia o total de eleitores, leia o numero de votos brancos, nulos e válidos e que mostre o percentual de cada categoria.

total_eleitores= int(input("Informe o total de eleitores:"))
brancos= int(input("Informe os votos brancos:"))
nulos= int(input("Informe os votos nulos:"))
validos= int(input("Informe os votos validos:"))

perc_brancos= (brancos/total_eleitores)*100
perc_nulos= (nulos/total_eleitores)*100
perc_validos= (validos/total_eleitores)*100

print(f"O percentual de votos brancos é {perc_brancos}%")
print(f"O percentual de votos nulos é {perc_nulos}%")
print(f"O percentual de votos validos é {perc_validos}%")
```

```
#Antecessor e Sucessor=
# Escreva um programa que: Leia um número inteiro e mostre seu antecessor e sucessor.

numero= int(input("Informe um numero:"))
antecessor= numero-1
sucessor= numero+1

print(f"O antecessor é {antecessor}")
print(f"O sucessor é {sucessor}")
```

```
#Cálculo do IMC=
# Escreva um programa que: Leia nome, peso(kg) e altura(m) de uma pessoa e que calcule e mostre o IMC.

nome= input("Informe o nome:")
peso= float(input("Informe o peso:"))
altura= float(input("Informe a altura:"))

imc= peso/(altura**2)
print(f"{nome},seu IMC é {imc}")
```
