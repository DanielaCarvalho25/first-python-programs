```
nome= input("Digite seu nome:")
peso= float(input("Digite seu peso(kg):"))
altura= float(input("Digite sua altura(m):"))
imc= peso/(altura**2)

print("Nome:",nome)
print("IMC:",imc)
if imc<23:
  print("Abaixo do peso")
elif imc>=23 and imc<=27:
  print("Peso normal")
elif imc>=28 and imc<=32:
  print("Sobrepeso")
elif imc>=33 and imc<=40:
  print("Obesidade")

else:
  print("IMC fora das faixas listradas")
```

```
nome= input("Digite o nome do produto:")
preco= float(input("Digite o preco do produto:"))
categoria= input("Digite a categoria (Perecível ou nao perecível):")

if categoria== "Perecível":
  preco_venda= preco+(preco*0.15)
else:
  preco_venda= preco+(preco*0.35)
print("Produto:",nome)
print("Preço de venda:",preco_venda)
```

```
nome= input("Digite o nome do cliente:")
tipo= input("Digite o tipo de cliente (Comercial ou Residencial):")
minutos= int(input("Digite a quantidade de minutos utilizados:"))

if tipo== "Residencial":
  conta = minutos * 0.70
else:
  if minutos<= 100:
    conta = minutos * 0.35
  else:
    conta = (100 * 0.35) + ((minutos - 100) * 0.50)
print("Cliente:",nome)
print("Tipo:", tipo)
print("Valor da conta", conta)
```

```
nome= input("Digite o nome do hóspede:")
idade= int(input("Digite a idade do hóspede:"))
dias= int(input("Digite a quantidade de dias:"))
tipo= input("Digite o tipo de apartamento (A, B, C, OU D):")
mes_aniversario = int(input("Digite o mês do seu aniversário (1 a 12):"))
mes_estadia = int(input("Digite o mês da sua estadia (1 a 12):"))
frigobar = float(input("Digite o valor gasto no frigobar:"))
diarias= int(input("Digite a quantidade de diarias:"))

#Definir valor da diária
if tipo== "A":
  diarias = 150
elif tipo== "B":
  diarias = 270
elif tipo== "C":
  diarias = 320
else:
  diarias = 410

#Calculas valor das diarias
total_diarias = diarias * dias

#Desconto no frigobar (aniversario)
if mes_aniversario == mes_estadia:
  frigobar = frigobar - (frigobar * 0.10)

#soma total
total = total_diarias + frigobar

#Desconto por idade
if idade > 60:
  total = total - (total * 0.30)

#Saída
print("Nome:",nome)
print("Tipo de apartamento:",tipo)
print("Valor da diarias:", total_diarias)
print("Valor gasto no frigobar:",frigobar)
print("Valor total a pagar:",total)
```
